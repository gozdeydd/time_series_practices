# ETS Model

```python
from sktime.forecasting.model_selection import temporal_train_test_split
y = aapl_df['Ingresos']
y_train, y_test = temporal_train_test_split(y, test_size=8)
# we will try to forecast y_test from y_train
# plotting for illustration
plot_series(y_train, y_test, labels=["y_train", "y_test"])
print(y_train.shape[0], y_test.shape[0])

y_test
```

```python
y_train, y_test = temporal_train_test_split(y = aapl_df['Ingresos'].astype('float64').to_period('Q'), test_size=8)
# we will try to forecast y_test from y_train
# plotting for illustration
plot_series(y_train, y_test, labels=["y_train", "y_test"])
print(y_train.shape[0], y_test.shape[0])
```

```python
# step 2: specifying forecasting horizon
fh = np.arange(1, 17)

# step 3: specifying the forecasting algorithm
aapl_auto_model = AutoETS(auto=True, sp=4, n_jobs=-1)
```

```python
aapl_auto_model.fit(y_train)
```

```python
print(aapl_auto_model.summary())
```

```python
# step 5: querying predictions for ETS Model
aapl_pred = aapl_auto_model.predict(fh)
print(aapl_pred)
```

```python
aapl_pred_ints = aapl_auto_model.predict_interval(fh, coverage=0.9)
aapl_pred_ints
```

```python

# optional: plotting predictions and past data
plot_series(y_train, aapl_pred,y_test, labels=["AAPL", "AAPL pred", "AAPL REAL"])
```

```python
fig, ax = plot_series(y_train, aapl_pred, y_test, labels=["AAPL", "AAPL pred", "AAPL REAL"])
ax.fill_between(
    ax.get_lines()[-2].get_xdata(),
    aapl_pred_ints[('Coverage', 0.9, 'lower')],
    aapl_pred_ints[('Coverage', 0.9, 'upper')],
    alpha=0.2,
    color=ax.get_lines()[-2].get_c(),
    label=f"90% prediction intervals",
)
ax.legend(loc='upper left')
```

```python
plot_series(y_train["2013":], aapl_pred,y_test, labels=["AAPL", "AAPL pred", "AAPL REAL"])
```

```python
fig, ax = plot_series(y_train["2013":], aapl_pred, y_test, labels=["AAPL", "AAPL pred", "AAPL REAL"])
ax.fill_between(
    ax.get_lines()[-2].get_xdata(),
    aapl_pred_ints[('Coverage', 0.9, 'lower')],
    aapl_pred_ints[('Coverage', 0.9, 'upper')],
    alpha=0.2,
    color=ax.get_lines()[-2].get_c(),
    label=f"90% prediction intervals",
)
ax.legend(loc='upper left');
```

# ARIMA

```python
aapl_df = pd.read_csv('./aapl_ok.csv', sep =';', decimal = ',')      

aapl_df['Ingresos'].astype('float64')
aapl_df['fecha'] = pd.to_datetime(aapl_df['fecha'])

aapl_df = aapl_df.set_index('fecha')

aapl_df
```

```python
# Plot Data
# Use seaborn style defaults and set the default figure size


sns.set(rc={'figure.figsize':(11, 4)})
ax = aapl_df.plot(marker='o', linestyle='-')
ax.set_ylabel('Ingresos Apple');
```

```python
# Resample to Quarterly I
aapl_df= aapl_df.resample("q").last()
aapl_df.tail()
```

```python
# Resample to Quarterly II
# SKtime format
aapl_df_q=aapl_df['Ingresos'].astype('float64').to_period('Q').sort_index()
aapl_df_q.tail()
```

```python
# Apply Log Transformer

transformer = LogTransformer()
log_aapl_df= transformer.fit_transform(aapl_df_q)
log_aapl_df.tail()
```

```python
# Plot Log Data
ax = log_aapl_df.plot(marker='o', linestyle='-')
ax.set_ylabel('Ingresos Apple')
ax.set_title('Ingresos Apple: Transformacion LOG')
```

```python
# Plot Log Data
fig, ax =plot_series(log_aapl_df, labels=["Ingresos"])
ax.set_title('Ingresos Apple: TransformaciÃ³n LOG')
```

```python
# Plot Log Data & Original Data
fig, ax =plot_series(aapl_df_q, labels=["Ingresos"])
ax.set_title('Ingresos Apple: Serie Original')
fig, ax =plot_series(log_aapl_df, labels=["Ingresos"])
ax.set_title('Ingresos Apple: TransformaciÃ³n LOG')
```

```python
# Autocorrelation Fuction Original Time Series
plot_correlations(log_aapl_df)
```

```python
# Autocorrelation Fuction
# d=1; D=0; S=4
# 
transf_diff=Differencer(lags=[1])
plot_correlations(transf_diff.fit_transform(log_aapl_df))
```

```python
# Autocorrelation Fuction
# d=0; D=1; S=4
# 
transf_diff=Differencer(lags=[4])
plot_correlations(transf_diff.fit_transform(log_aapl_df))
```

```python
# Autocorrelation Fuction
# d=1; D=1; S=4
# 
transf_diff=Differencer(lags=[1,4])
plot_correlations(transf_diff.fit_transform(log_aapl_df))
```

```python
# Sktime fucntions
# Forecast horizon and Split function 
from sktime.forecasting.base import ForecastingHorizon
from sktime.forecasting.model_selection import temporal_train_test_split
```

```python
# Select Train & test sample
# we will try to forecast y_test from y_train
# plotting for illustration
# original and log samples
y_train, y_test = temporal_train_test_split(y =aapl_df_q, test_size=8)
log_y_train, log_y_test = temporal_train_test_split(y =log_aapl_df, test_size=8)
plot_series(y_train, y_test, labels=["Train", "Test"])
# Time Series Size
print(y_train.shape[0], y_test.shape[0])
```

```python
# Forecast Horizon
fh = np.arange(len(y_test)) + 1  # forecasting horizon
fh
```

```python
# Sktime Auto ARIMA Function
from sktime.forecasting.arima import AutoARIMA
```

```python
#  Auto ARIMA Model

forecaster = AutoARIMA(sp=4,suppress_warnings=True)
forecaster.fit(log_y_train)
```

```python
#  Auto ARIMA Model Summary
print(forecaster.summary())
```

```python
#  Auto ARIMA Forecast
log_y_pred = forecaster.predict(fh)
log_y_pred
```

```python
#  Auto ARIMA Forecast
# Original Time series (Invert log transformation)
np.exp(log_y_pred)
```

```python
# Forecast Accuracy
# MAPE
mean_absolute_percentage_error(log_y_test, log_y_pred)
```

```python
# Forecast Accuracy
# MAPE 
# Orignal Time Serie

mean_absolute_percentage_error(y_test, np.exp(log_y_pred))
```

```python
# Forecast Accuracy
#Mean Squared Error
rmse = MeanSquaredError(square_root=True)
rmse(log_y_test, log_y_pred)
```

```python
# Forecast Accuracy
#Mean Squared Error
rmse = MeanSquaredError(square_root=True)
rmse(y_test, np.exp(log_y_pred))
```

```python
# Forecast Accuracy Plot
# Plotting predictions and past data
plot_series(y_train, np.exp(log_y_pred),y_test, labels=["AAPL", "AAPL pred", "AAPL REAL"])
```

```python
# Forecast Accuracy Plot
# Plotting predictions and past data
# Zoom 2019 -2021
plot_series(y_train["2019":], np.exp(log_y_pred),y_test, labels=["AAPL", "AAPL pred", "AAPL REAL"])
```
